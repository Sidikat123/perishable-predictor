Promotion Effectiveness

```python
# Discount Effectiveness

# Merge datasets so we can access prices and discount info
sales_merged = weekly_sales.merge(product_details, on="Product_ID")

# Create discount analysis summary
discount_analysis = (
    sales_merged.groupby("Discount_Percent")
    .agg(Total_Units_Sold=('Units_Sold', 'sum'))
    .reset_index()
)

# Calculate sales lift vs no discount
no_discount_sales = discount_analysis.loc[
    discount_analysis["Discount_Percent"] == 0, "Total_Units_Sold"
].values[0]

discount_analysis["Sales_Lift_vs_No_Discount (%)"] = (
    (discount_analysis["Total_Units_Sold"] - no_discount_sales) / no_discount_sales
) * 100

# Sort for clean plotting
discount_analysis = discount_analysis.sort_values("Discount_Percent")

# Visualize (line chart)
plt.figure(figsize=(8, 5))
sns.lineplot(data=discount_analysis, x="Discount_Percent", y="Total_Units_Sold", marker="o", color="blue")
plt.title("Discount Effectiveness: Units Sold vs Discount (%)")
plt.xlabel("Discount Percentage")
plt.ylabel("Total Units Sold")
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(8, 5))
sns.barplot(data=discount_analysis, x="Discount_Percent", y="Sales_Lift_vs_No_Discount (%)", color="orange")
plt.title("Sales Lift vs No Discount (%) by Discount Level")
plt.xlabel("Discount Percentage")
plt.ylabel("Sales Lift (%)")
plt.axhline(0, color='gray', linestyle='--')
plt.tight_layout()
plt.show()
```

**Promotion Effectiveness**
- Units sold are highest at 0% discount and decline as discounts increase
- Every discount tier shows a negative sales lift vs. no discount.

**Insights**
- Blanket discounting is not driving volume which may likely be applied to slow movers items or at the wrong times.
- Discounts are eroding margin without lifting sales, so there is need to prioritize targeted, data-led promotions (e.g. Stock Keeping Unit specific, seasonal, or inventory-risk based).
- Test smaller, time-boxed discounts, bundles, or loyalty offers instead of broad cuts which measure lift by cohort to confirm ROI.

**Marketing Spend Effectiveness**

```python
sales_merged = weekly_sales.merge(product_details, on="Product_ID", how="left")

# Keep rows with marketing spend > 0 to avoid divide-by-zero and noise
df = sales_merged[sales_merged["Marketing_Spend"] > 0].copy()

# Metrics
corr = df["Marketing_Spend"].corr(df["Units_Sold"])
avg_cost_per_unit = df["Marketing_Spend"].sum() / df["Units_Sold"].sum()

# Correlation visualization: scatter + simple trendline
x = df["Marketing_Spend"].values
y = df["Units_Sold"].values
m, b = np.polyfit(x, y, 1)   # least-squares line

plt.figure(figsize=(7,5))
plt.scatter(x, y, alpha=0.4)
plt.plot(np.sort(x), m*np.sort(x)+b)   # trendline
plt.title("Marketing Spend vs Units Sold (Correlation Visual)")
plt.xlabel("Marketing Spend (£)")
plt.ylabel("Units Sold")
plt.text(0.02, 0.95, f"Correlation = {corr:.3f}", transform=plt.gca().transAxes)
plt.tight_layout()
plt.show()


# KPI visualization: average marketing cost per unit sold
plt.figure(figsize=(6,3.5))
plt.axis("off")
plt.title("Average Marketing Cost per Unit Sold")
plt.text(0.5, 0.5, f"£{avg_cost_per_unit:.2f} per unit",
         ha="center", va="center", fontsize=18)
plt.tight_layout()
plt.show()
```

**Marketing Spend Analysis**
- Marketing spend has near-zero correlation with units sold (r ≈ 0.04), and the average marketing cost is £0.25 per unit which suggest current spend is not materially driving volume.

**Key insights for decisions:**
- Treat £0.25/unit as a guardraili.e. only fund campaigns where incremental margin > £0.25 per unit.
- Re-allocate/stop low-ROI campaigns and concentrate budget on channels, SKUs, and regions that show measurable lift.
- Run controlled A/B tests (by SKU/region/season) to quantify true incremental lift before scaling.
- Link spend to inventory risk (e.g. short-life Bakery) and proven price elasticity, not blanket promotions.

```python
# Calculate product performance metrics
product_perf = sales_merged.groupby('Product_ID').agg(
    Total_Units_Sold=('Units_Sold', 'sum'),
    Total_Revenue=('Price', lambda x: (x * sales_merged.loc[x.index, 'Units_Sold']).sum()), # Calculate total revenue
    Total_Wastage=('Wastage_Units', 'sum')
).reset_index()

# Add product name and category to product_perf
product_perf = product_perf.merge(product_details[['Product_ID', 'Product_Name', 'Product_Category']], on='Product_ID')

# Filter out products with less than 1000 units sold for wastage analysis
product_perf_filtered = product_perf[product_perf['Total_Units_Sold'] >= 1000].copy()

display(product_perf.head())
display(product_perf_filtered.head())
```

```python
# Draw Top-5 vertical bars on a given Axes (no new figure)
def draw_top5_vertical(ax, df, value_col, title, ylabel,
                       rotation=35, fmt="{:,.0f}", top_n=5, ylim=None):
    df5  = df.sort_values(value_col, ascending=False).head(top_n)
    names = df5["Product_Name"].tolist()
    vals  = df5[value_col].tolist()

    bars = ax.bar(names, vals)
    for bar, v in zip(bars, vals):  # value labels above bars
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
                fmt.format(v), ha="center", va="bottom", fontsize=9)

    ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.tick_params(axis="x", labelrotation=rotation)
    if ylim is not None:
        ax.set_ylim(0, ylim)

# --- 1 row, 3 columns ---
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# 1) Top 5 best-selling products (units)
draw_top5_vertical(
    axes[0], product_perf, "Total_Units_Sold",
    title="Top 5 Best-Selling Products",
    ylabel="Total Units Sold",
    rotation=35
)

# 2) Top 5 revenue-generating products (£)
draw_top5_vertical(
    axes[1], product_perf, "Total_Revenue",
    title="Top 5 Revenue-Generating Products",
    ylabel="Total Revenue (£)",
    rotation=35, fmt="£{:,.0f}"
)

# 3) Top 5 highest-wastage products (>1,000 units sold)
draw_top5_vertical(
    axes[2], product_perf_filtered, "Total_Wastage",
    title="Top 5 Highest-Wastage Products (>1,000 units sold)",
    ylabel="Total Wastage Units",
    rotation=35
)

plt.tight_layout()
plt.show()
```

```python
# --- Build monthly_perf ---
weekly_sales["Month"] = pd.to_datetime(
    weekly_sales["Week_Number"].astype(str) + "-1",
    format="%Y-W%W-%w"
).dt.to_period("M")

monthly_perf = (weekly_sales
    .groupby("Month", as_index=False)
    .agg(Total_Units_Sold=("Units_Sold", "sum"),
         Total_Wastage=("Wastage_Units", "sum"),
         Total_Marketing_Spend=("Marketing_Spend", "sum"))
)
monthly_perf["Marketing_Efficiency"] = (
    monthly_perf["Total_Marketing_Spend"] / monthly_perf["Total_Units_Sold"]
)
monthly_perf["Wastage_Rate (%)"] = (
    monthly_perf["Total_Wastage"]
    / (monthly_perf["Total_Units_Sold"] + monthly_perf["Total_Wastage"]) * 100
)

# --- Sort & x-axis labels ---
monthly_sorted = monthly_perf.sort_values("Month").copy()
x = monthly_sorted["Month"].astype(str).values

# --- 2x2 subplots ---
fig, axes = plt.subplots(2, 2, figsize=(14, 8))

# (1) Monthly Sales
axes[0, 0].plot(x, monthly_sorted["Total_Units_Sold"], marker="o")
axes[0, 0].set_title("Monthly Sales Trend")
axes[0, 0].set_xlabel("Month")
axes[0, 0].set_ylabel("Total Units Sold")

# (2) Monthly Wastage (units)
axes[0, 1].plot(x, monthly_sorted["Total_Wastage"], marker="o", color="red")
axes[0, 1].set_title("Monthly Wastage Trend")
axes[0, 1].set_xlabel("Month")
axes[0, 1].set_ylabel("Total Wastage Units")

# (3) Total_Marketing_Spend
axes[1, 0].plot(x, monthly_sorted["Total_Marketing_Spend"], marker="o", color="orange")
axes[1, 0].set_title("Monthly Marketing_Spend Trend")
axes[1, 0].set_xlabel("Month")
axes[1, 0].set_ylabel("Total Marketing_Spend")


# (4) Marketing Efficiency
axes[1, 1].plot(x, monthly_sorted["Marketing_Efficiency"], marker="o", color="green")
axes[1, 1].set_title("Marketing Efficiency by Month (Spend per Unit)")
axes[1, 1].set_xlabel("Month")
axes[1, 1].set_ylabel("£ per Unit Sold")

# Shared formatting
for ax in axes.ravel():
    ax.tick_params(axis="x", labelrotation=45)
    ax.grid(True)

plt.tight_layout()
plt.show()
```

**Seasonal Trends:**
- Sales are cyclical with peaks in Jan, Apr, Jul, and Oct
- Wastage is highest in Jan and Apr and lowest around Aug
- Marketing spend spikes in Jan and Dec
- However, spend-per-unit (efficiency) is worst in Jun and Dec and best in Feb and Nov.

**Key insights & actions:**
- Post-holiday hangover: Jan shows high sales and high wastage. Hence, the need to tighten forecasting and bake smaller batches for Bakery and accelerate markdowns early in Jan.
- Summer trough: Aug has the cleanest waste profile. There is need to maintain lean inventory and avoid blanket promotions.
- Marketing ROI: Despite heavy spend in Dec, efficiency is poorest. There is need to reallocate toward proven channels/SKUs, apply tighter targeting and shorter flights.
- Replicate winners: Feb and Nov deliver best efficiency.There is need to reuse those audience, channel, and creative mixes.
- Budget timing: Shift budget from Jun and Dec due to low efficiency and move it to months with higher natural demand i.e. Apr and Jul. Nevertheless, there is need to control inventory so as to avoid repeat of waste spikes.

**Shelf Life Impact**

```python
shelf_life_analysis = sales_merged.groupby("Product_Category").agg(
    Avg_Shelf_Life_Days=('Shelf_Life_Days', 'mean'),
    Total_Wastage=('Wastage_Units', 'sum')
).reset_index()

display(shelf_life_analysis)
```

```python
# Expecting shelf_life_analysis with columns:
# Product_Category, Avg_Shelf_Life_Days, Total_Wastage

# Use one consistent order (by shortest shelf life first)
order = (shelf_life_analysis
         .sort_values("Avg_Shelf_Life_Days", ascending=True)["Product_Category"]
         .tolist())
df = (shelf_life_analysis
      .set_index("Product_Category").loc[order].reset_index())

fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# 1: Average shelf life (days)
axes[0].barh(df["Product_Category"], df["Avg_Shelf_Life_Days"])
axes[0].set_title("Average Shelf Life (Days) by Category")
axes[0].set_xlabel("Days")
axes[0].set_ylabel("Product Category")
for i, v in enumerate(df["Avg_Shelf_Life_Days"].round(1)):
    axes[0].text(v, i, f"{v:.1f}", va="center", ha="left")

# 2: Total wastage units
axes[1].barh(df["Product_Category"], df["Total_Wastage"])
axes[1].set_title("Total Wastage Units by Category")
axes[1].set_xlabel("Wastage Units")
for i, v in enumerate(df["Total_Wastage"].values):
    axes[1].text(v, i, f"{int(v):,}", va="center", ha="left")

plt.tight_layout()
plt.show()

# 3: Relationship: shelf life vs wastage (scatter with labels)
plt.figure(figsize=(7, 5))
plt.scatter(shelf_life_analysis["Avg_Shelf_Life_Days"], shelf_life_analysis["Total_Wastage"])
for _, r in shelf_life_analysis.iterrows():
    plt.annotate(r["Product_Category"],
                 (r["Avg_Shelf_Life_Days"], r["Total_Wastage"]),
                 textcoords="offset points", xytext=(5, 3))
plt.title("Shelf Life vs Wastage")
plt.xlabel("Avg Shelf Life (Days)")
plt.ylabel("Total Wastage Units")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
nominal_columns = ['Product_Category', 'Region']  # one hot
high_cardinal = ['Product_Name', 'Supplier_Name']  # mean encoding
```

```python
from sklearn.preprocessing import OneHotEncoder

# Copy original data
encode_data = merged_data

# OneHot Encoding for nominal columns
encoded_columns = []
ohe = OneHotEncoder(sparse_output=False, dtype=int)

for col in nominal_columns:
    transformed_col = ohe.fit_transform(encode_data[[col]])
    column_names = [f"{col}_{cat}" for cat in ohe.categories_[0]]
    transformed_df = pd.DataFrame(transformed_col, columns=column_names, index=encode_data.index)
    encoded_columns.append(transformed_df)

# Concatenate all encoded nominal columns
encode_data = pd.concat([encode_data] + encoded_columns, axis=1)
encode_data.drop(columns=nominal_columns, inplace=True)

# Mean Encoding for high-cardinality columns using 'Units_Sold' as target
for col in high_cardinal:
    encode_data[col] = encode_data.groupby(col)['Units_Sold'].transform('mean')
```

```python
# Convert to datetime columns

merged_data['Week_Number'] = pd.to_datetime(merged_data['Week_Number'] + '-1', format='%G-W%V-%u')

merged_data['Week_Number']
```
